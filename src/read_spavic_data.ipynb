# read excel data
read an xlsx file using pandas and make chart from the data

## read and plot temperature data

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import time

station = [
    'Bekasi',
    'Sudirman',
    'Manggarai',
    'Jakarta Kota',
    'Tanjung Periok',
    'Taman Kota',
    'Tanah Abang'
]

df = pd.read_excel("../data/spavic/temperature.xlsx")

x = [i.strftime("%H:%M") for i in df['Time']]

y1 = df[station[0]]
y2 = df[station[1]]
y3 = df[station[2]]
y4 = df[station[3]]
y5 = df[station[4]]
y6 = df[station[5]]
y7 = df[station[6]]

plt.figure(figsize=(8, 5))

plt.plot(x, y1, linestyle='-', marker='.', color='r')
plt.plot(x, y2, linestyle='--', marker='o', color='g')
plt.plot(x, y3, linestyle='-.', marker='v', color='b')
plt.plot(x, y4, linestyle='-', marker='^', color='k')
plt.plot(x, y5, linestyle='--', marker='x', color='m')
plt.plot(x, y6, linestyle='-.', marker='o', color='c')
plt.plot(x, y7, linestyle='--', marker='.', color='y')

plt.ylim(20, 40)
plt.title("Stations Temperature in a day")
plt.xlabel("Time")
plt.xticks(rotation = 90)
plt.ylabel("Temperature")

plt.legend(station, loc='upper right', ncol=3)
plt.grid()

plt.show()
```
